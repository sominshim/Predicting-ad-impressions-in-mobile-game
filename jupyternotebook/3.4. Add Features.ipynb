```python
skin_col = ['boy', 'doctor', 'firefighter', 'police', 'pringles', 'starbucks', 'suit', 'worker']
asset['skin_cnt'] = asset[skin_col].sum(axis=1)
```

```python
asset.drop(['boy', 'doctor', 'firefighter', 'police', 'pringles', 'starbucks', 'suit', 'worker'], axis=1, inplace=True)
asset.rename(columns={'0':'default_skin_cnt'}, inplace=True)
```

```python
asset.fillna(0, inplace=True)
asset.head(2)
```

```python
asset['coin_cnt'] = asset[['in_coin_cnt', 'out_coin_cnt']].sum(axis=1)
asset['coin_sum'] = asset[['in_coin_sum', 'out_coin_sum']].sum(axis=1)
asset['coin_minmax_diff'] = asset.curr_var_max - asset.curr_var_min
```

```python
asset['upgrade_coin_sum'] = asset[['upgrade_atk_var_sum', 'upgrade_atk_speed_var_sum',
                                   'upgrade_coin_var_sum', 'upgrade_hp_var_sum']].sum(axis=1)
asset['rv_coin_sum'] = asset[['rv_collection_coin_var_sum', 'rv_result_bonus_coin_var_sum', 
                              'rv_result_coin_var_sum']].sum(axis=1)
```

### 5. 전체 out 재화량 대비 능력치 업그레이드에 쓴 재화량
- 'upgrade_coin_sum' / 'out_coin_sum'

### 6. 전체 in 재화량 대비 리워드 비디오로 얻은 재화량

```python
def outlier_iqr(data, column): 
    print('--------- '+column+'피처--------- ')
    # lower, upper 글로벌 변수 선언하기     
#     global lower, upper    
    
    # 4분위수 기준 지정하기     
    q25 = np.quantile(data[column], 0.25) 
    q75 = np.quantile(data[column], 0.75)          
    
    # IQR 계산하기     
    iqr = q75 - q25    
    # outlier cutoff 계산하기     
    cut_off = iqr * 1.5          
    
    # lower와 upper bound 값 구하기     
    lower = q25 - cut_off  
    upper = q75 + cut_off 
    
    print('IQR : ',iqr)     
    print('lower bound : ', lower)     
    print('upper bound : ', upper)    
    
    # 1사 분위와 4사 분위에 속해있는 데이터 각각 저장하기     
    data1 = data[data[column] > upper]     
    data2 = data[data[column] < lower]    
    
    print('총 이상치 개수 : ', data1.shape[0] + data2.shape[0])
    print('해당 데이터에서 이상치의 비율 : {}%'.format(round((data1.shape[0] + data2.shape[0])/data.shape[0]*100,2)))
    
    # 이상치 데이터 index 반환
    outlier_id = [i for i in (data1.ad_id + data2.ad_id)]
    
    return data[~data['ad_id'].isin(outlier_id)]
```
